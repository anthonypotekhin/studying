```python
from statistics import mean # программа, вычисляющая total sum sqares для набора данных Х

x = ([3, 1, 3, 1, 3, 13])

def tss(a):
    total = 0
    for i in a:
        total = total + ((i-mean(a))**2)
    return total
print(tss(x))
```
