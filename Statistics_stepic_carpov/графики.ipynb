```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

t = sp.Symbol('t') # задаем независимую переменную
function_x = sp.sympify('t**2-t') # определеляем функцию (зависимую переменную) 
# функция sympify.sp.sympify(expression) преобразует параметрическое выражение в общее математическое выражение.
print(function_x)

interval = np.arange(-10, 12, 1) # задаем интервал значений для независимой переменной
# Функция np.arange(start, stop, step) создает массив NumPy со значениями 
# в интервале (start,stop) с приращением шага, то есть мы идем от start до stop-1 с определенным шагом step.

print(interval)

x_values = [function_x.subs(t, value) for value in interval]

fig = plt.figure(figsize=(7, 7))
ax = plt.axes()
# fig.add_axes([0, 100, 0, 100])
ax.plot(x_values)

ax = plt.gca()

# plot X - axis    
ax.axhline(y=0, color='k')

# plot Y - axis    
ax.axvline(x=0, color='k')

plt.show()
```

```python
from math import *
from tkinter import *

f = input('f(x):')

root = Tk()

canv = Canvas(root, width = 1000, height = 1000, bg = "white")
canv.create_line(500,1000,500,0,width=2,arrow=LAST) 
canv.create_line(0,500,1000,500,width=2,arrow=LAST) 

First_x = -500;

for i in range(16000):
	if (i % 800 == 0):
		k = First_x + (1 / 16) * i
		canv.create_line(k + 500, -3 + 500, k + 500, 3 + 500, width = 0.5, fill = 'black')
		canv.create_text(k + 515, -10 + 500, text = str(k), fill="purple", font=("Helvectica", "10"))
		if (k != 0):
			canv.create_line(-3 + 500, k + 500, 3 + 500, k + 500, width = 0.5, fill = 'black')
			canv.create_text(20 + 500, k + 500, text = str(k), fill="purple", font=("Helvectica", "10"))
	try:
		x = First_x + (1 / 16) * i
		new_f = f.replace('x', str(x))
		y = -eval(new_f) + 500
		x += 500
		canv.create_oval(x, y, x + 1, y + 1, fill = 'black')
	except:
		pass
canv.pack()	
root.mainloop()
```
