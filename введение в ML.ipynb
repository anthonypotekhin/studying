```python
# можно фильтровать и по неравенству - лишь бы оно давало булево значение:

import pandas as pd
import numpy as np
students_Performance= pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')

mean_writing_score = students_Performance['writing score'].mean()   # берем среднее значение столбца writing score

display(students_Performance.loc[students_Performance['writing score']> mean_writing_score])    # локализуем только те строки, 
                                                                                                # у которых значение столбца
                                                                                                # writing score больше среднего
```

```python
# фильтрация по нескольким условиям
# возможна только при условии правильного синтаксиса запроса 
# (подробнее тут https://stackoverflow.com/questions/36921951/truth-value-of-a-series-is-ambiguous-use-a-empty-a-bool-a-item-a-any-o) 
# 1) объединяй условия через оператор в виде значка 
# 2) каждое условие забирай в скобки

import pandas as pd
import numpy as np
students_Performance= pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')

mean_writing_score = students_Performance['writing score'].mean()   # берем среднее значение столбца writing score
# прописываем в имя переменной Запроса два ОДНОЗНАЧНЫХ булевых значения (балл больше среднего, пол только женский)
zapros = (students_Performance['writing score'] > mean_writing_score) & (students_Performance.gender== 'female')

# фильтруем (локализуем) таблицу по заданному запросу
display(students_Performance.loc[zapros])  
```

```python
# переименуем столбцы (уберем пробелы нахрен в названии столбцов, чтобы можно было обращаться к ним через точку)

import pandas as pd
import numpy as np
students_Performance= pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')

students_Performance = students_Performance\
                                    .rename(columns=
                                            {'parental level of education': 'parental_level_of_education',
                                            'test preparation course': 'test_preparation_course',
                                            'math score': 'math_score',
                                            'reading score': 'reading_score',
                                            'writing score': 'writing_score'})

students_Performance.head()
```

```python
# проверим фильтрацию более компакнтым способом:

students_Performance.loc[students_Performance.writing_score > 74].head()
```

```python
# либо используем встроенный метод query():

students_Performance.query('writing_score > 74').head()     # в кавычках внутри аргумента указываем условие фильтра

```

```python
# при этом если я ищу булево значение в фильтре для значений, которые заполнены типов данных строка, то нужны доп кавычки:

students_Performance.query("gender == 'female' ").head() 
```

```python
# возможна также комбинация запросов: 

students_Performance.query("gender == 'female' & math_score >75 ") 
```

```python
# случай, когда нужно фильтровать по значению, которого мы еще не знаем, но он вычисляется в процессе исполнения:

some_score = 78             # допустим, в процессе вычисления некое значение получилось равным 78
```

```python
# тогда, чтобы отфильтровать по нашему полученному значению должно быть:
students_Performance.query("math_score > @some_score ") 
```

```python
# дополнительная удобная фильтрация, если колонок очень много, а нам нужны только те, что в названии содержат 'score'
# 
# тогда удобно воспользоваться тем фактом, что в списке по dataFrame содержатся имена заголовков колонок:

print(list(students_Performance))   # выведет список имен колонок  

# а теперь фильтруем нашу таблицу так, чтобы вывести только те столбцы, в которых есть score

display(students_Performance[ [i for i in list(students_Performance) if 'score' in i]].head() )

```

```python
# метод для фильтрации колонок, который намного проще - встроенный метод filter():

students_Performance.filter(like= 'score').head()
```

```python
# кстати, отвлечемся - посчитать число ТОЛЬКО строк можно воспользовавшись функцией len():

print(len(students_Performance))
```

```python
# дополнительно методу filter можно передать значение оси - направление по столбикам или строкам: 
# удобно запоминать r0ws (строки)= 0, co1umns = 1 (столбцы) 

# теперь фильтрация может вестись либо по строкам, либо по столбцам:

display(with_names.filter(like='e', axis=1))        # выводим только те столбцы, в названии которых есть буква 'e'

display(with_names.filter(like='e', axis=0))        # выводим только те строки, в названии которых есть буква 'e'


display(with_names.filter(like='e'))        # по умолчанию axis принимает значение 1 (только для столбцов)
```

```python
# сгруппируем данные по столбцу пола и возьмем средние значения других столбцов (к которым применимо):

students_Performance.groupby('gender').mean()
```

```python
# но правильнее указывать конкретные требования к группируемым значениям dataFrame:

students_Performance.groupby('gender').aggregate({'math_score': 'mean', 'reading_score': 'mean', 'writing_score':'mean'})
```

```python
# для удобства можно переименовать столбцы с измененными значениями

students_Performance.groupby('gender', as_index= False)\
    .aggregate({'math_score': 'mean', 'reading_score': 'mean', 'writing_score':'mean'})\
        .rename(columns= {'math_score': 'mean_math', 'reading_score': 'mean_reading', 'writing_score':'mean_writing'})
```

```python
# а если нужна подгруппировка? 
# возьмем за основную группировку пол, а в качестве подгруппировки выступит национальность 

mean_scores = students_Performance.groupby(['gender', 'race/ethnicity'])\
    .aggregate({'math_score': 'mean', 'reading_score': 'mean', 'writing_score':'mean'})\
        .rename(columns= {'math_score': 'mean_math', 'reading_score': 'mean_reading', 'writing_score':'mean_writing'})

mean_scores
```
