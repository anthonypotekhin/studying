### можно создать новый столбец, значения в котором будут сформированы из значений существующей таблицы

```python
# дописать новый столбец очень просто:
# объявляем dataFrame[new_name] = серия, которую хотим записать в эту колонку 
# то есть по сути логика та же самая как в словаре - ты по заданному ключу формируешь новое значение, которое хочешь дописать, 
# а если такого ключа нет, то он будет создан и к нему допишется новое значение (в данном случае Серия)

# например, хотим добавить столбец суммы всех баллов:

students_Performance['total_score'] = \
    students_Performance.math_score + \
        students_Performance.reading_score + \
            students_Performance.writing_score

students_Performance        # наш датафрейм перезаписан с учетом нового столбца так, 
                            # что каждый элемент серии соответствующего столбца складывается с указанными

# если я не хочу трогать существующий dataFrame , поэтому все вышеперечисленное я проделаю с его копией:

# with_total_scores = students_Performance.copy(deep= True)

# with_total_scores['total_score'] = \
#    with_total_scores.math_score + \
#        with_total_scores.reading_score + \
#            with_total_scores.writing_score

# with_total_scores
```

```python
# что делать, если я хочу добавить или  преобразовать сразу НЕСКОЛЬКО столбцов? 
# метод assign(new_column = значение серии, new_col2 = значение серии, ....)

students_Performance =  students_Performance.assign(tot_score_log = np.log(students_Performance.total_score))

# обрати внимание, что пришлось перезаписать dataFrame! 

students_Performance

# в этом примере мы берем уже измененную таблицу, добавляем новый столбец, присваиваем ему имя и говорим, 
# что за значения должны там храниться - в данном случае логарифм от значений серии в столбце total_score
```

```python
vodorosli = pd.read_csv('http://stepik.org/media/attachments/course/4852/algae.csv')

display(vodorosli)

mean_concentrations = vodorosli.groupby('genus')\
    .aggregate({'sucrose':'mean', 'oleic_acid':'mean', 'alanin':'mean', 'citrate':'mean', 'glucose':'mean'} )\
        .rename({'sucrose':'mean_suc', 'oleic_acid':'mean_oleic', 'alanin':'mean_al', 'citrate':'mean_cit', 'glucose':'mean_gluc'} )

mean_concentrations
```

```python

# 1) ищем дисперсию концентрации цитрата в трех видах выдорослей
var_concentration = vodorosli.groupby('group')\
    .aggregate({'sucrose':'var', 'oleic_acid':'var', 'alanin':'var', 'citrate':'var', 'glucose':'var'} )\
        .rename(columns = {'sucrose':'suc_var', 'oleic_acid':'oleic_var', 'alanin':'al_var', 'citrate':'cit_var', 'glucose':'gluc_var'} )

display(var_concentration)

# 2) ищем число видов бурых, зеленых и красных водорослей

num_species = vodorosli.groupby(['group']).nunique()
display(num_species)



# 3) ищем размах (разницу между максимальным и минимальным значением) сахарозы во всех видах водорослей
# используем встроенную функцию, которая считает разницу максимального и минимального значения - numpy.ptp
range_concetrate = vodorosli.groupby('group')\
    .aggregate({'sucrose': np.ptp })\
        .rename(columns = {'sucrose':'suc_range'})

display(range_concetrate)

# другой способ - тупо написать функцию самому
def range_elems(spis):
    return max(spis)-min(spis)

range_concetrate1 = vodorosli.groupby('group')\
    .aggregate({'sucrose': range_elems })\
        .rename(columns = {'sucrose':'suc_range'})

display(range_concetrate1)
```

### ЗАДАЧА: построить распределения по параметрам ирисов

скачайте датасэт со значениями параметров ирисов, постройте их распределения и отметьте правильные утверждения, глядя на график. 
Распределение должно быть по всем образцам, без разделения на вид.

полезные ссылки: https://seaborn.pydata.org/generated/seaborn.kdeplot.html

https://seaborn.pydata.org/generated/seaborn.distplot.html

```python
iris_df= pd.read_csv('https://stepik.org/media/attachments/course/4852/iris.csv')

iris_df
```

где хранятся заголовки столбцов? 

если воспринимать dataframe как словарь, то можно обратиться к заголовкам как к ключам:

```python
print(iris_df.keys())
print(iris_df.keys()[0])

print(list(iris_df.keys()))         # укажем тип данных наверняка - как список (тогда с ними проще работать)
```

построим гистограмму частот по каждому из параметров: 

- в качестве параметров возьмем только те столбцы, в которых содержится информация о параметрах ирисов

- пройдемся по ним циклом

```python
# возьмем только те столбцы, в чьих именах содержатся параметры и пройдемся по ним циклом
for param in [p for p in list(iris_df.keys()) if 'length' in p or 'width' in p]: 
    display(iris_df[param].hist())


# можно еще проще пойти циклом по столбцам - всего лишь ' for column in df: '
```

получилось не оч красиво, поэтому воспользуемся kdeplot() для построения гистограмм частот:


```python
iris_df_params = iris_df.drop(['Unnamed: 0', 'species'], axis=1) #  предварительно уберем столбцы, которые мешаются

for column in iris_df_params:
    sns.kdeplot(iris_df_params)                                 # прогоним циклом по столбцам
```

выводы, которые можно сделать глядя на совмещенный график гистограмм:
- параметры petal length и petal width имеют бимодальную форму распределения (по двер вершины)
- параметры sepal length и widgth имеют унимодальную форму
- наибольший размах (разница между наибольшим и наименьшим значением) имеет параметр petal width (потому что ее кривая самая растянутая вдоль оси х, которая содержит как раз всю область значений параметра)

НАПЕЧАТАТЬ ИМЕНА ОСЕЙ

```python
ax_hists = sns.kdeplot(iris_df_params) 

ax_hists.set_xlabel('param_value\n(величина параметра)')
ax_hists.set_ylabel('Rate\n(частота)')
```
