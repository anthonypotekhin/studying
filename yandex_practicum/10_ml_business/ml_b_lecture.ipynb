Конверсии

Процесс продаж можно описать схемой: посетитель → продажа → покупатель.
Всё начинается с того, что в компанию обращается посетитель. Визит на сайт, телефонный звонок, установка мобильного приложения или посещение магазина — не важно. Главное, что человек заинтересовался продуктами или услугами компании. 


Потом происходит продажа: посетитель добавляет товар в корзину, проходит обучающие уровни в игре или общается с продавцом. Хотя эти процессы различаются, цель у них одна — превратить посетителя в покупателя. Чем эффективнее у бизнеса получается делать это, тем лучше идут дела. Поэтому одна из самых важных метрик — это конверсия. Она показывает, какая доля посетителей совершила целевое действие (например, скачала приложение или купила подписку).

Если в магазин зашли 1000 человек, а 10 из них что-то купили, то конверсия посетителей в покупателей равна 10/1000 = 0.01 или 1%.

### A/B тесты 

Суть: 

A/B-тестирование, или сплит-тестирование (англ. A/B testing; Split testing, от англ. «разделять») — техника проверки гипотез. Позволяет оценить, как изменение сервиса или продукта повлияет на пользователей. Проводится так: аудиторию делят на две группы — контрольную (A) и тестовую (В). Группа A видит начальный сервис, без изменений. Группа B получает новую версию, которую и нужно протестировать. 

Эксперимент длится фиксированное время (например, две недели). В ходе тестирования собираются данные о поведении пользователей в разных группах. Если ключевая метрика в тестовой группе выросла по сравнению с контрольной, новую функциональность внедряют.

Прежде чем провести A/B-тест, часто делают проверку корректности — A/A тест. В нём пользователей делят на две контрольные группы, которые видят одинаковый сервис. Ключевая метрика у групп отличаться не должна: если она разная — ищите ошибку.

Терминология A/B-тестирования

- Контроль (группа контроля, контрольная группа) — это группа пользователей, для которых не вводят изменений. В нашем примере — это пользователи, которые увидят старый вариант чекаута.

- Тест (группа теста, тестовая группа) — это группа пользователей, для которых вводят тестируемое изменение. В нашем примере — это пользователи, которые увидят новый экран чекаута.
- Тритмент (англ. treatment — лечение) — это изменение, которое вводят для тестовой группы. В нашем случае это вид экрана чекаута.
- Фича (англ. feature — особенность) — похожий по значению на «тритмент», но более общий термин, который используется в IT-компаниях для обозначения нового функционала в продукте.
- Раскатить фичу — ввести изменение для всех пользователей или для части.
- Метрика — показатель, изменение которого анализируют по результатам эксперимента. Например, прибыль с пользователя или количество заказов.
- Зелёный тест — эксперимент, по результатам которого зафиксировали статистически значимый прирост метрики.
- Красный тест — эксперимент, по результатам которого зафиксировали статистически значимое падение метрики.
- Серый тест — эксперимент, по результатам которого не зафиксировали статистически значимого изменения метрики.

Стат значимость A/B теста 

Работа выполняется через сравнение средних значений метрик

Нулевая гипотеза H₀ звучит так: «Новая функциональность не улучшает метрики». Тогда альтернативная гипотеза H₁ такая: «Новая функциональность улучшает метрики».

одностороннюю альтернативную гипотезу принимают, если проверяемое значение либо намного больше, либо меньше принятого в нулевой гипотезе. Нас интересует отклонение только в одну сторону — знак «больше». 

Пример:

Интернет-магазин мармеладных червячков «Ползучая тянучка» разместил анимированного червяка-маскота на сайте. Нововведение «ползает» по экрану пользователя, пока тот выбирает товар. 

Посмотрите на две выборки средних чеков в интернет-магазине до и после запуска червячка. Напечатайте на экране значения средних до и после внедрения анимированного маскота. 

Проверьте гипотезу, что средний чек увеличился. Выберите уровень значимости, равный 5%. Выведите на экран значение p-value и результат проверки гипотезы.

```python
import pandas as pd

sample_before = pd.Series([
    436, 397, 433, 412, 367, 353, 440, 375, 414, 
    410, 434, 356, 377, 403, 434, 377, 437, 383,
    388, 412, 350, 392, 354, 362, 392, 441, 371, 
    350, 364, 449, 413, 401, 382, 445, 366, 435,
    442, 413, 386, 390, 350, 364, 418, 369, 369, 
    368, 429, 388, 397, 393, 373, 438, 385, 365,
    447, 408, 379, 411, 358, 368, 442, 366, 431,
    400, 449, 422, 423, 427, 361, 354])

sample_after = pd.Series([
    439, 518, 452, 505, 493, 470, 498, 442, 497, 
    423, 524, 442, 459, 452, 463, 488, 497, 500,
    476, 501, 456, 425, 438, 435, 516, 453, 505, 
    441, 477, 469, 497, 502, 442, 449, 465, 429,
    442, 472, 466, 431, 490, 475, 447, 435, 482, 
    434, 525, 510, 494, 493, 495, 499, 455, 464,
    509, 432, 476, 438, 512, 423, 428, 499, 492, 
    493, 467, 493, 468, 420, 513, 427])

print("Cреднее до:", sample_before.mean())
print("Cреднее после:", sample_after.mean())


# критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
alpha = 0.05

from scipy import stats as st

results = st.ttest_ind(sample_before, sample_after, alternative='less')

# тест односторонний: p-value будет в два раза меньше
pvalue = results.pvalue

print('p-значение: ', pvalue)

if pvalue < alpha:
    print("Отвергаем нулевую гипотезу: скорее всего средний чек увеличился")
else:
    print("Не получилось отвергнуть нулевую гипотезу: скорее всего средний чек не увеличился")
```

Суть: надергать рзных выборок, получить из них интересующее значение, построить распределение



Бутстреп применим для любых выборок. Это полезно, когда:
- Наблюдения не описываются нормальным законом;

- Для искомых величин нет статистических тестов.

В действительности не стоит всегда рассчитывать на нормальное распределение. Разберём технику бутстрепа для оценки доверительного интервала и анализа результатов A/B-теста.

#### Бутстреп для анализа A/B-теста

Как работает на примере.

Предположим, нам нужно проверить изменилась ли средняя сумма в чеке до изменений и после.

Тогда:
- возьмем две выборки, которые содержат информацию о сумме в чеке до и после изменений
- посчитаем в каждом среднее, получим разницу между средними
- объединим выборки
- из общей выборки в цикле "надергаем" с помощью бутстрепа несколько новых выборок того же размера и разделим их ровно пополам для каждой итерации
- на каждой итерации посчитаем разницу между средними значениями половинок выборок
- соберем все эти средние в контейнер

Теперь зададим нулевую и альтернативную гипотезы:

- Нулевая предполагает равенство средних чеков в обеих группах. 
- Альтернативная — в экспериментальной группе средний чек выше. 

Найдём p-value - оно  равно доле превышений значений


```python
import pandas as pd
import numpy as np

# данные контрольной группы A
samples_A = pd.Series([
     98.24,  97.77,  95.56,  99.49, 101.4 , 105.35,  95.83,  93.02,
    101.37,  95.66,  98.34, 100.75, 104.93,  97.  ,  95.46, 100.03,
    102.34,  98.23,  97.05,  97.76,  98.63,  98.82,  99.51,  99.31,
     98.58,  96.84,  93.71, 101.38, 100.6 , 103.68, 104.78, 101.51,
    100.89, 102.27,  99.87,  94.83,  95.95, 105.2 ,  97.  ,  95.54,
     98.38,  99.81, 103.34, 101.14, 102.19,  94.77,  94.74,  99.56,
    102.  , 100.95, 102.19, 103.75, 103.65,  95.07, 103.53, 100.42,
     98.09,  94.86, 101.47, 103.07, 100.15, 100.32, 100.89, 101.23,
     95.95, 103.69, 100.09,  96.28,  96.11,  97.63,  99.45, 100.81,
    102.18,  94.92,  98.89, 101.48, 101.29,  94.43, 101.55,  95.85,
    100.16,  97.49, 105.17, 104.83, 101.9 , 100.56, 104.91,  94.17,
    103.48, 100.55, 102.66, 100.62,  96.93, 102.67, 101.27,  98.56,
    102.41, 100.69,  99.67, 100.99])

# данные экспериментальной группы B
samples_B = pd.Series([
    101.67, 102.27,  97.01, 103.46, 100.76, 101.19,  99.11,  97.59,
    101.01, 101.45,  94.8 , 101.55,  96.38,  99.03, 102.83,  97.32,
     98.25,  97.17, 101.1 , 102.57, 104.59, 105.63,  98.93, 103.87,
     98.48, 101.14, 102.24,  98.55, 105.61, 100.06,  99.  , 102.53,
    101.56, 102.68, 103.26,  96.62,  99.48, 107.6 ,  99.87, 103.58,
    105.05, 105.69,  94.52,  99.51,  99.81,  99.44,  97.35, 102.97,
     99.77,  99.59, 102.12, 104.29,  98.31,  98.83,  96.83,  99.2 ,
     97.88, 102.34, 102.04,  99.88,  99.69, 103.43, 100.71,  92.71,
     99.99,  99.39,  99.19,  99.29, 100.34, 101.08, 100.29,  93.83,
    103.63,  98.88, 105.36, 101.82, 100.86, 100.75,  99.4 ,  95.37,
    107.96,  97.69, 102.17,  99.41,  98.97,  97.96,  98.31,  97.09,
    103.92, 100.98, 102.76,  98.24,  97.  ,  98.99, 103.54,  99.72,
    101.62, 100.62, 102.79, 104.19])

# фактическая разность средних значений в группах
AB_difference = samples_B.mean() - samples_A.mean()
print("Разность средних чеков:", AB_difference)

alpha = 0.05
    
state = np.random.RandomState(12345)

bootstrap_samples = 1000
count = 0
for i in range(bootstrap_samples):
    # объедините выборки
    united_samples = pd.concat([samples_A, samples_B]) 

    # создайте подвыборку
    subsample = united_samples.sample(frac=1, replace=True, random_state=state)

    us_len =  united_samples.__len__()
    
    # разбейте подвыборку пополам
    subsample_A = subsample.iloc[: int(us_len/2)]
    subsample_B = subsample.iloc[int(us_len/2) :]

    # найдите разницу средних
    bootstrap_difference = subsample_B.mean() - subsample_A.mean()

    
    # если разница не меньше фактической, увеличиваем счётчик
    if bootstrap_difference >= AB_difference:
        count += 1

# p-value равно доле превышений значений
pvalue = 1. * count / bootstrap_samples
print('p-value =', pvalue)

if pvalue < alpha:
    print("Отвергаем нулевую гипотезу: скорее всего, средний чек увеличился")
else:
    print("Не получилось отвергнуть нулевую гипотезу: скорее всего, средний чек не увеличился")
```

#### Бутстреп для моделей

Как с помощью Bootstrap оценить доверительные интервалы для моделей машинного обучения?

Школа экспресс-занятий по английскому языку «Заспúкать за 5 минут» разрабатывает модель для оценки вероятности — придёт ученик на занятие или нет. 

Каждый день поступает много заявок. Приоритет выставляется по очерёдности: кто раньше записался — того и занятие. Около половины учеников не приходят и не оплачивают урок. Руководство школы решило оставить занятия только с теми учениками, кто приходит вероятнее всего. Из-за возможных репутационных рисков компания внедрит это нововведение, если есть гарантия заметного увеличения выручки. Чтобы принять верное решение, оцените распределение вероятности для прибыли.

Вот важные условия задачи:
- Модель для предсказания вероятности посещения урока уже обучена. Предсказания находятся в файле eng_probabilities.csv, а правильные ответы — в eng_target.csv.
- Одно занятие стоит 1000 рублей. В день можно запланировать до 10 уроков. Текущая выручка за день — 5000 рублей (половина учеников отменяет занятие).
- За день поступает в среднем 25 заявок.

- Достаточно большая выручка для внедрения — 7500 рублей. Её вероятность должна быть не меньше 99%

Для начала напишем  функцию revenue() (англ. «выручка»), которая подсчитывает и возвращает выручку. 

Она получает на вход:
- список ответов target — пришёл ли ученик на урок;
- список вероятностей probabilities — модель оценивает, придёт ученик или нет;
- сколько учеников посещает занятия за день count.

Функция должна выбрать учеников с наибольшей вероятностью посещения и на основе ответов подсчитать возможную выручку.

```python
import pandas as pd

def revenue(target, probabilities, count):
    only_came_students = target[probabilities.sort_values(ascending=False).index].iloc[:count]

    return 1000 * only_came_students.sum() # 1000 рублей прибыли за одного пришедшего ученика

target = pd.Series([1,   1,   0,   0,  1,    0])
probab = pd.Series([0.2, 0.9, 0.8, 0.3, 0.5, 0.1])

res = revenue(target, probab, 3)

print(res)
```

Теперь найдем  1%-квантиль выручки, выполнив процедуру Bootstrap с 1000 повторений. Почему именно 1% ? Потому что по условию задачи вероятность получения увеличения выручки должна составлять 99%

```python
# открываем файлы
# возьмём индекс '0', чтобы перевести данные в pd.Series
target = pd.read_csv('/Users/tony_builder/Documents/GitHub/studying/yandex_practicum/10_ml_business/data/eng_target.csv')['0']
probabilities = pd.read_csv('/Users/tony_builder/Documents/GitHub/studying/yandex_practicum/10_ml_business/data/eng_probabilities.csv')['0']

def revenue(target, probabilities, count):
    probs_sorted = probabilities.sort_values(ascending=False)
    selected = target[probs_sorted.index].iloc[:count]
    return 1000 * selected.sum()

state = np.random.RandomState(12345)
    
values = []
for i in range(1000):
    # отбираем итстинные выборки по 25 учеников
    target_subsample = target.sample(25,                # здесь задаем среднее число учеников в день - именно такую выборку будет формировать bootstrap
                                    replace=True,       # разрешаем повторения в выборках
                                    random_state=state) # задаем случайные наборы
    
    # отбираем предсказанную выборку с теми же индексами, что и истинные
    probs_subsample = probabilities[target_subsample.index]         # отбираем наиболее вероятные события

    revenue_value = revenue(target_subsample,                       # считаем значение выручки на каждой итерации
                            probs_subsample, 
                            10)                                     # значение выбрано =10, т.к максимальное число уроков в день может быть только 10
    
    values.append(revenue_value)

values = pd.Series(values)
lower = values.quantile(0.01)       # считаем 1% квантиль

mean = values.mean()
print("Средняя выручка:", mean)
print("1%-квантиль:", lower)
```

Вывод: 

Почти 10 000 рублей. Средняя выручка близка к максимально возможной. 

И с вероятностью 99% выручка может быть не менее 6000 р.  (по условию задачи нужно было 7500)

Голосование по большинству

Извлечение из нескольких разметок наиболее часто встречающегося варианта.

Допустим, есть три оценки врачей (label_1, label_2, label_3). Это "размеченные" тремя людьми данные.
ЧТобы принять наиболее релевантную - берем саую частую оценку (это и есть оценка большинства)

```python
import pandas as pd

data = pd.read_csv('/Users/tony_builder/Documents/GitHub/studying/yandex_practicum/10_ml_business/data/heart_labeled.csv')
data.head()

```

```python
# создаем новый столбец со значением самого часто повторящегося значения в трех столбцах для каждой строки

data['target']= data.apply(lambda row: int(Counter([row['label_1'], row['label_2'], row['label_3']]).most_common(1)[0][0]), 
                           axis=1)

data.head()
```

Оценка качества на кросс-валидации

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

features = data.drop(['target', 'label_1', 'label_2', 'label_3'], axis=1)
target = data['target']

model = DecisionTreeClassifier(random_state=0)

scores = cross_val_score(model, features, target, cv=5) 

final_score = pd.Series(scores).mean()

print('Средняя оценка качества модели:', final_score)
```
