```python
# —Ñ—É–Ω–∫—Ü–∏—è –ø–æ–ª—É—á–µ–Ω–∏—è –∏–º–µ–Ω–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π

def get_var_name(var, caller_locals):
    import inspect
    for name, value in caller_locals.items():
        if value is var:
            return name
    return None
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –∏ boxplot –¥–ª—è –ª—é–±–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ –ª—é–±–æ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞

def plot_hist_boxplot(some_df, col, col_name, dataset_name):

    df_for_plot= some_df[col]

    # –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –∫–≤–∞—Ä—Ç–∏–ª–∏
    q25, q75 = df_for_plot.quantile(0.25), df_for_plot.quantile(0.75)
    min_val = int(q25- 1.5*(q75-q25))
    max_val= int(q75 +  1.5*(q75-q25))

    # –°–æ–∑–¥–∞–µ–º –Ω–æ–≤—ã–π –≥—Ä–∞—Ñ–∏–∫
    plt.figure(figsize=(10, 10))

    # –°—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É
    plt.subplot(2, 1, 1)  # 2 —Å—Ç—Ä–æ–∫–∏, 1 —Å—Ç–æ–ª–±–µ—Ü, –≥—Ä–∞—Ñ–∏–∫ 1
    plt.hist(df_for_plot, bins=20, range=(min_val, max_val))
    plt.title(f'–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –¥–ª—è –ø—Ä–∏–∑–Ω–∞–∫–∞ {col_name} –¥–ª—è –¥–∞—Ç–∞—Å–µ—Ç–∞ {dataset_name}')
    plt.xlabel(f'–∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–∞ {col_name}')
    plt.ylabel(f'–ß–∞—Å—Ç–æ—Ç–∞')


    # –ü–æ–º–µ—á–∞–µ–º –∫–≤–∞—Ä—Ç–∏–ª–∏ –Ω–∞ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–µ
    plt.axvline(q25, color='r', linestyle='--', label='25th percentile')
    plt.axvline(q75, color='g', linestyle='--', label='75th percentile')
    plt.legend()

    # –°—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ boxplot
    plt.subplot(2, 1, 2)  # 2 —Å—Ç—Ä–æ–∫–∏, 1 —Å—Ç–æ–ª–±–µ—Ü, –≥—Ä–∞—Ñ–∏–∫ 2
    plt.boxplot(df_for_plot, vert=False)
    plt.title(f'Boxplot –ø—Ä–∏–∑–Ω–∞–∫–∞ {col_name} –¥–ª—è –¥–∞—Ç–∞—Å–µ—Ç–∞ {dataset_name}')
    plt.xlabel(f'–ó–Ω–∞—á–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∞ {col_name}')

    # –ü–æ–º–µ—á–∞–µ–º –∫–≤–∞—Ä—Ç–∏–ª–∏ –Ω–∞ boxplot
    plt.axvline(q25, color='r', linestyle='--', label='25th percentile')
    plt.axvline(q75, color='g', linestyle='--', label='75th percentile')
    plt.legend()


    # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –≥—Ä–∞—Ñ–∏–∫–∏
    plt.tight_layout()  # –î–ª—è –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –≤—ã—Ä–∞–≤–Ω–∏–≤–∞–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
    plt.show()


        
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è —Å—Ç–æ–ª–±—á–∞—Ç–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã

def barplot_df(dataset, col, col_name, dataset_name):
    dataset[col].value_counts().plot(kind='bar', legend=True, figsize=(10,5))
    plt.title(f'–î–∏–∞–≥—Ä–∞–º–º–∞ —á–∞—Å—Ç–æ—Ç—ã –ø—Ä–∏–∑–Ω–∞–∫–∞ {col_name} –¥–ª—è –¥–∞—Ç–∞—Å–µ—Ç–∞ {dataset_name}')
    plt.xlabel(f'–∑–Ω–∞—á–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∞ {col_name}')
    plt.ylabel(f'–ß–∞—Å—Ç–æ—Ç–∞ –ø—Ä–∏–∑–Ω–∞–∫–∞ {col_name}')
    plt.show()

    
```

```python
def pie_plot(dataset, col, dataset_name):
    plt.title(f'–ö—Ä—É–≥–æ–≤–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞ —á–∞—Å—Ç–æ—Ç—ã –ø—Ä–∏–∑–Ω–∞–∫–∞ {col} –¥–ª—è –¥–∞—Ç–∞—Å–µ—Ç–∞ {dataset_name}')

    palette_color = sns.color_palette('dark')

    # –±–ª–æ–∫ –ø–æ–ª—É—á–µ–Ω–∏—è —Å–∞–º–æ–≥–æ –ø–æ–ø—É–ª—è—Ä–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
    feat_counts = dataset[col].value_counts()
    # –ü–æ–ª—É—á–µ–Ω–∏–µ —Å–∞–º–æ–≥–æ —á–∞—Å—Ç–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
    most_common_feat = feat_counts.idxmax()


    # Plotting data on chart
    patches, texts, autotexts = plt.pie(feat_counts, 
                                        labels=feat_counts.index, 
                                        explode=(feat_counts.index == most_common_feat) * 0.1,      # –≤–∑—Ä—ã–≤–∞–µ–º –¥–∏–∞–≥—Ä–∞–º–º—É - –≤—ã—Ç–∞—Å–∫–∏–≤–∞–µ–º —Å–∞–º–æ–µ –ø–æ–ø—É–ª—è—Ä–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
                                        autopct='%1.1f%%', 
                                        startangle=140,
                                        colors=palette_color,
                                        textprops={'color':'white'})

    # –ù–∞—Å—Ç—Ä–æ–π–∫–∞ —Ü–≤–µ—Ç–∞ —Ç–µ–∫—Å—Ç–∞ –Ω–∞ —Ñ–æ–Ω–µ –≤–æ–∫—Ä—É–≥ –ø–∏—Ä–æ–≥–∞
    for text in texts:
        text.set_color('black')
    
    plt.show()
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –ø–æ–ª—É—á–µ–Ω–∏—è –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏—Ö –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –ª—é–±–æ–π —Ç–∞–±–ª–∏—Ü—ã

def analysis_df(some_df: pd.DataFrame, cat_list: list, num_list: list):
    # –±–ª–æ–∫ –≤—ã—Ç—è–≥–∏–≤–∞–Ω–∏—è –∏–º–µ–Ω–∏
    import inspect
    caller_locals = inspect.currentframe().f_back.f_locals

    dataset_name = get_var_name(some_df, caller_locals)
    

    print_red(f'–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ —Ç–∞–±–ª–∏—Ü—ã {dataset_name}')  # –ø–µ—á–∞—Ç–∞–µ–º –∏–º—è —Ç–∞–±–ª–∏—Ü—ã (–∏–∑ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π)
    display(some_df.describe().T)

    # –±–ª–æ–∫ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º –∏ –¥–∏–∞–≥—Ä–∞–º–º

    for col in num_list:
        plot_hist_boxplot(some_df=some_df, col_name=col, col=col, dataset_name=dataset_name)

    for col in cat_list:   # –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
        try:
            barplot_df(dataset=some_df, col_name=col, col=col, dataset_name=dataset_name)
            pie_plot(dataset=some_df, col=col, dataset_name=dataset_name)
        except KeyError:
            continue
```

```python
# One-hot encoding

try:

    ohe_pipe = Pipeline(
        [
            (
                'simpleImputer_ohe', 
                SimpleImputer(missing_values=np.nan, strategy='most_frequent')
            ),
            (
                'ohe', 
                OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
            )
        ]
    ) 

except:
        ohe_pipe = Pipeline(
        [
            (
                'simpleImputer_ohe', 
                SimpleImputer(missing_values=np.nan, strategy='most_frequent')
            ),
            (
                'ohe', 
                OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False)
            )
        ]
    ) 

# Ordinal encoding
ord_pipe = Pipeline(
    [
        (
            'simpleImputer_before_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',  
            OrdinalEncoder(
                categories=[
                    ['junior', 'middle', 'sinior'], 
                    ['low', 'medium', 'high'],
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
        (
            'simpleImputer_after_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)

"""num_pipe = Pipeline(
    [
        (
            'simpleImputer_num', 
            SimpleImputer(missing_values=np.nan, strategy='mean')
        ),
        (
            'scaler', 
            MinMaxScaler()
        )
    ]
)"""

num_pipe = Pipeline(
    [
        (
            'simpleImputer_num', 
            SimpleImputer(strategy='median')
        ),
        (
            'scaler', 
            MinMaxScaler()
        )
    ]
)


# –û–±—â–∏–π –ø–∞–π–ø–ª–∞–π–Ω
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, cat_cols+bin_cols),
        ('ord', ord_pipe, ordinal_cols),
        ('num', num_pipe, num_cols)
    ], 
    remainder='passthrough'
) 
```

#### –®–∞–≥ 5. –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

–°–æ–∑–¥–∞–¥–∏–º –∏—Ç–æ–≥–æ–≤—ã–π –ø–∞–π–ø–ª–∞–π–Ω –ø–æ —Å–ª–µ–¥—É—é—â–µ–º—É –ø—Ä–∏–Ω—Ü–∏–ø—É:
- –±–∞–∑–æ–≤–∞—è –º–æ–¥–µ–ª—å - –ª–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è
- –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–∞—è –º–æ–¥–µ–ª—å (–ø–æ–¥–±–æ—Ä —Å –ø–æ–º–æ—â—å—é GridSearch) - –¥–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π

```python
RANDOM_STATE = 42
```

```python
pipe_final = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', LinearRegression())
    ]
)
```

```python
param_grid  = [
    {
        'models': [LinearRegression(n_jobs=-1)],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_leaf_nodes': range(5, 400, 15),
        'models__min_samples_leaf': range(1, 10, 2),
        'models__min_samples_split': range(2, 10, 2),
        'models__max_depth': range(2, 150, 15),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    }
]
```

–ó–∞–¥–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é –ø–æ–ª—É—á–µ–Ω–∏—è –º–µ—Ç—Ä–∏–∫–∏ SMAPE 

<font color='red'>–ö—Ä–∏—Ç–µ—Ä–∏–π —É—Å–ø–µ—Ö–∞: SMAPE ‚â§15 –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ.</font>



```python
def smape(y_true, y_pred):
    return 100*np.mean(2*np.abs(y_pred - y_true) / (np.abs(y_pred) + np.abs(y_true)))

# –°–æ–∑–¥–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏
smape_scorer = make_scorer(smape, greater_is_better=False)
```

–ó–∞–¥–∞–¥–∏–º –ø–æ–¥–±–æ—Ä—â–∏–∫ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤

```python
# –°–æ–∑–¥–∞–Ω–∏–µ –ø–æ–¥–±–æ—Ä—â–∏–∫–∞ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_distributions=param_grid, 
    scoring=smape_scorer, 
    cv=5,
    random_state=RANDOM_STATE,
    n_jobs=-1
)
```

- 40 000 —É –º–∏–¥–ª–∞
- 60 000 —É —Å–µ–Ω–∏–æ—Ä–∞

```python
def split_salary(s):
    if s<25_000: return '<25k'
    elif 25_000<=s<40_000: return '25-40k' 
    elif 40_000<=s<55_000: return '40-55k' 
    elif 55_000<=s<70_000: return '55-70k'
    else: return '>70k' 


target_feat = 'quit'

# –§–∏–ª—å—Ç—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ —Ü–µ–ª–µ–≤–æ–º—É –ø—Ä–∏–∑–Ω–∞–∫—É - –æ—Å—Ç–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ —Ç–µ—Ö, –∫—Ç–æ —É–≤–æ–ª–∏–ª–∏—Å—è
filtered_data = train_quit_df.query('quit == "yes"')
filtered_data['employment_years']= filtered_data.loc[:, 'employment_years'].apply(lambda x: '>4 years' if x>3 else str(x))
filtered_data.salary = filtered_data.salary.apply(split_salary)

# –°–æ–∑–¥–∞–µ–º —Å–ø–∏—Å–æ–∫ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
features = train_quit_df.drop('quit', axis=1).columns.to_list()

# –°–æ–∑–¥–∞–µ–º –ø–æ–¥–≥—Ä–∞—Ñ–∏–∫–∏ —Å –ø–æ–º–æ—â—å—é Seaborn
fig, axes = plt.subplots(2, 4, figsize=(16, 10))
plt.suptitle('–ü–æ—Ä—Ç–µ—Ç —É–≤–æ–ª–∏–≤—à–µ–≥–æ—Å—è —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–∞')

# –ü—Ä–æ—Ö–æ–¥–∏–º—Å—è –ø–æ –≤—Å–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∞–º –∏ —Å—Ç—Ä–æ–∏–º –∫—Ä—É–≥–æ–≤—ã–µ –¥–∏–∞–≥—Ä–∞–º–º—ã
for i, feature in enumerate(features):
    row = i // 4
    col = i % 4
    
    
    
    # –ì—Ä—É–ø–ø–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ –ø—Ä–∏–∑–Ω–∞–∫—É –∏ –ø–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ–º –∏—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ
    counts = filtered_data[feature].value_counts()
    
    # –û–ø—Ä–µ–¥–µ–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –±—É–¥–µ—Ç –≤–∑—Ä—ã–≤–∞—Ç—å—Å—è (–Ω–∞–∏–±–æ–ª—å—à–µ–µ)
    explode_index = counts.idxmax()
    explode = [0.1 if idx == explode_index else 0 for idx in counts.index]
    
   # –°—Ç—Ä–æ–∏–º –∫—Ä—É–≥–æ–≤—É—é –¥–∏–∞–≥—Ä–∞–º–º—É —Å –ø–æ–º–æ—â—å—é Matplotlib –∏ —Ü–≤–µ—Ç–æ–≤–æ–π —Å—Ö–µ–º–æ–π –∏–∑ Seaborn
    patches, texts, autotexts = axes[row, col].pie(counts, 
                       labels=counts.index, 
                       explode=explode, 
                       autopct='%1.1f%%', 
                       startangle=140, 
                       colors=sns.color_palette('dark')
                        )
    
    axes[row, col].set_title(feature)

    # –ù–∞—Å—Ç—Ä–æ–π–∫–∞ —Ü–≤–µ—Ç–∞ —Ç–µ–∫—Å—Ç–∞ –≤–Ω—É—Ç—Ä–∏ –ø–∏—Ä–æ–≥–∞ –Ω–∞ –±–µ–ª—ã–π
    for autotext in autotexts:
        autotext.set_color('white')




plt.tight_layout()
plt.show()
```

–ò—Ç–∞–∫, —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ = 0 - –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫ –Ω–µ —É–≤–æ–ª–∏–ª—Å—è, –∞ =1 - —á—Ç–æ —É–≤–æ–ª–∏–ª—Å—è

–≥–æ—Ç–æ–≤–∏–º —Ñ–∏–Ω–∞–ª—å–Ω—ã–π –ø–∞–π–ø–ª–∞–π–Ω:
- –ø–æ —Å—É—Ç–∏ –¥–ª—è –∫–æ–¥–∏—Ä–æ–≤–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –æ–Ω –Ω–µ –∏–∑–º–µ–Ω–∏–ª—Å—è (–¥–æ–±–∞–≤–∏–ª—Å—è –µ—â–µ –æ–¥–∏–Ω —Å—Ç–æ–ª–±–µ—Ü)
- –Ω—É–∂–Ω–æ –≤–Ω–µ—Å—Ç–∏ –∏–∑–º–µ–Ω–µ–Ω–∏—è –ø–æ –º–æ–¥–µ–ª—è–º - —Å–µ–π—á–∞—Å —Ä–µ—à–∞–µ–º –∑–∞–¥–∞—á—É –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏

```python
# –û–±—â–∏–π –ø–∞–π–ø–ª–∞–π–Ω
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, cat_cols+bin_cols),
        ('ord', ord_pipe, ordinal_cols),
        ('num', num_pipe, num_cols_quit)
    ], 
    remainder='passthrough'
) 
```

```python
pipe_final = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', LogisticRegression(random_state=RANDOM_STATE))
    ]
)
```

–ó–∞–¥–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –ø–æ–∏—Å–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏

```python
param_grid = [
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(3, 10),
        'models__min_samples_leaf': range(3, 10),
        'models__min_samples_split': range(2, 10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ KNeighborsClassifier() 
    {
       'models': [KNeighborsClassifier()],
        # —É–∫–∞–∑—ã–≤–∞–µ–º –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä –º–æ–¥–µ–ª–∏ n_neighbors
       'models__n_neighbors': range(2,5),
        # —É–∫–∞–∑—ã–≤–∞–µ–º —Å–ø–∏—Å–æ–∫ –º–µ—Ç–æ–¥–æ–≤ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è
       'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },

    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ LogisticRegression()
    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE,
            solver='saga' 
            )],
        'models__C': range(1,5),
        'models__penalty': ['l1', 'l2'],
        
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },

    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ SVC()
    {
        'models': [SVC(kernel = 'poly') 
                    ],
        'models__degree': range(2,6),
        
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    }
]



```

–ó–∞–ø—É—Å–∫–∞–µ–º –ø–æ–∏—Å–∫

```python
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
```

```python
randomized_search.fit(X_train, y_train)
```

```python
print('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –∏ –µ—ë –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:\n\n', randomized_search.best_estimator_)
print ('–ú–µ—Ç—Ä–∏–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:', randomized_search.best_score_)

# –ø—Ä–æ–≤–µ—Ä—å—Ç–µ —Ä–∞–±–æ—Ç—É –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
# —Ä–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ –ø—Ä–æ–≥–Ω–æ–∑ –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
preds= randomized_search.predict_proba(X_test)
print(f'–ú–µ—Ç—Ä–∏–∫–∞ ROC-AUC –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ: {roc_auc_score(y_test_enc, preds[:,1])}')
```
