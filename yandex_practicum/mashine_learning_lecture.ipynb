### Машинное обучение

#### Урок 1. Моделирование и машинное обучение

Машинное обучение всегда идёт по одной и той же схеме:

1) Подготовка учебного материала — эталонных данных. (разметка)
2) Обучение, тестирование и запуск модели для предсказаний. 
3) Поддержка готовой модели, чтобы она сохраняла точность.

Схема работы с Моделью машиннного обучения



    Существует три вида машинного обучения: с учителем, без учителя и с подкреплением

##### Обучение с учителем 

Самый простой вариант научить машину делать выводы — показать ей правильные ответы столько раз, чтобы она их выучила. Компьютер сможет легко узнавать леопардов, если загрузить достаточно картинок и отметить, где среди них именно эти звери. 

Так работает обучение с учителем: 

- Модель сразу получает весь учебный датасет — признаковое описание объекта.
- В признаковом описании объекта есть столбец с целевым признаком — например названием животного.
- Значения целевого признака указаны. <font color='red'>Заранее размеченные данные — это и есть учитель.</font>

По правильным примерам, то есть с учителем, компьютер можно научить решать четыре класса задач: регрессии, классификации, ранжирования и рекомендации. 

<font color='red'>**Регрессия**</font> Популярная задача по прогнозу числовых показателей. Пример регрессии — предсказание стоимости квартиры по количеству комнат, метражу, удалённости от центра города и другим параметрам. 

<font color='red'>**Классификация**</font>. Ещё одна частая задача — отнести объект к одному из классов. Например, для модерации текстов: они либо содержат запрещённую лексику, либо нет. 

<font color='red'>**Ранжирование**</font>. Если компьютер должен расставить объекты в каком-то порядке — значит, это задача ранжирования. Типичный пример — работа поисковой системы. Любой поисковик определяет, какую веб-страницу показать в выдаче первой, какую — второй… То есть присваивает им ранг — ранжирует. 

<font color='red'>**Рекомендация**</font>. Задача рекомендации тесно связана с ранжированием, но они разные. Ранжирование отвечает на прямой запрос пользователя и выстраивает варианты ответов в лучшем порядке. Рекомендация позволяет предложить пользователю то, о чëм он не спрашивал, но чем может заинтересоваться.

##### Обучение с подкреплением

В обучении с учителем есть один минус. Как вы знаете, при таком подходе целевые признаки рассчитывают по их прежним значениям в связке с другими свойствами объекта. Из-за этого предсказания получаются статистически верными, но не учитывают изменения данных в процессе обучения модели. А вдруг мир за это время станет другим — какие тогда регрессии? К тому же как быть, если правильных решений задачи ещё нет? В таких случаях компьютеру приходится учиться без подсказок — методом проб и ошибок. Машина будет ориентироваться на результат своих действий, хороший или плохой. Такой способ называется обучением с подкреплением. Вот его свойства:

- Есть целевой признак — например прибыль от покупки и продажи акций. В этом смысле всё как в обучении с учителем.

- Нет заранее подготовленного датасета. Учебные данные поступают из среды, в которой учится модель. Пример такой среды — биржа.
- Нет самого учителя — то есть примеров правильных решений. Компьютер вырабатывает их самостоятельно. <font color='green'>Если модель МО — это робот для трейдинга, то она будет запоминать, какие её действия принесли прибыль, а какие — убыток</font>. Поэтому обучение — с подкреплением.

почему бы не обучить робота для трейдинга с помощью регрессии? Ведь он не первый на свете покупает и продаёт акции. Дело в том, что некоторые бизнес-цели требуют анализировать информацию сразу же после того, как она поступила. В торговле на бирже большую роль играют новости. Одни мировые события цикличны, а другие непредсказуемы и происходят один раз, но сильно влияют на рынок. Чтобы прогнозировать котировки акций как можно эффективнее, нужно постоянно обновлять учебные данные и выводы о верных действиях. Обучение с подкреплением не панацея, но регрессия и вовсе здесь бесполезна. Она учит по старым данным.

В обучении с подкреплением пока не выделили классы задач.

##### Обучение без учителя

В обучении без учителя машина ищет не целевой признак одного объекта, а любые закономерности на уровне объектов. Это полезно в задачах кластеризации и снижения размерности.

<font color='red'>Кластеризация</font>. Группировка объектов по их признакам. Модель сама решает по каким. 
С помощью кластеризации можно, например, распределить клиентов по предпочтениям и делать им подходящие предложения.

<font color='red'>Снижение размерности</font>. Бывает, что в признаковом описании указано слишком много свойств объекта — сотни, тысячи. Моделям машинного обучения тяжело делать выводы по такому количеству возможных переменных. Значит, лучше его уменьшить. Специалист по Data Science может сгруппировать признаки вручную, но задача снижения размерности помогает не тратить на это время.

Вот основные черты обучения без учителя:
- Целевого признака нет, потому что задача — выявить закономерности между объектами, а не признаками.

- Примеров правильных решений тоже нет. По ходу обучения компьютер делает выводы сам — без учителя.
- В отличие от обучения с подкреплением, все учебные данные есть изначально. Так же, как в обучении с учителем.

Обощенная схема моделей с учителем, подкреплением, без учителя
