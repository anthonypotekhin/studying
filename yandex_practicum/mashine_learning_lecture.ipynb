### Машинное обучение

#### Урок 1. Моделирование и машинное обучение

Машинное обучение всегда идёт по одной и той же схеме:

1) Подготовка учебного материала — эталонных данных. (разметка)
2) Обучение, тестирование и запуск модели для предсказаний. 
3) Поддержка готовой модели, чтобы она сохраняла точность.

Схема работы с Моделью машиннного обучения



    Существует три вида машинного обучения: с учителем, без учителя и с подкреплением

##### Обучение с учителем 

Самый простой вариант научить машину делать выводы — показать ей правильные ответы столько раз, чтобы она их выучила. Компьютер сможет легко узнавать леопардов, если загрузить достаточно картинок и отметить, где среди них именно эти звери. 

Так работает обучение с учителем: 

- Модель сразу получает весь учебный датасет — признаковое описание объекта.
- В признаковом описании объекта есть столбец с целевым признаком — например названием животного.
- Значения целевого признака указаны. <font color='red'>Заранее размеченные данные — это и есть учитель.</font>

По правильным примерам, то есть с учителем, компьютер можно научить решать четыре класса задач: регрессии, классификации, ранжирования и рекомендации. 

<font color='red'>**Регрессия**</font> Популярная задача по прогнозу числовых показателей. Пример регрессии — предсказание стоимости квартиры по количеству комнат, метражу, удалённости от центра города и другим параметрам. 

<font color='red'>**Классификация**</font>. Ещё одна частая задача — отнести объект к одному из классов. Например, для модерации текстов: они либо содержат запрещённую лексику, либо нет. 

<font color='red'>**Ранжирование**</font>. Если компьютер должен расставить объекты в каком-то порядке — значит, это задача ранжирования. Типичный пример — работа поисковой системы. Любой поисковик определяет, какую веб-страницу показать в выдаче первой, какую — второй… То есть присваивает им ранг — ранжирует. 

<font color='red'>**Рекомендация**</font>. Задача рекомендации тесно связана с ранжированием, но они разные. Ранжирование отвечает на прямой запрос пользователя и выстраивает варианты ответов в лучшем порядке. Рекомендация позволяет предложить пользователю то, о чëм он не спрашивал, но чем может заинтересоваться.

##### Обучение с подкреплением

В обучении с учителем есть один минус. Как вы знаете, при таком подходе целевые признаки рассчитывают по их прежним значениям в связке с другими свойствами объекта. Из-за этого предсказания получаются статистически верными, но не учитывают изменения данных в процессе обучения модели. А вдруг мир за это время станет другим — какие тогда регрессии? К тому же как быть, если правильных решений задачи ещё нет? В таких случаях компьютеру приходится учиться без подсказок — методом проб и ошибок. Машина будет ориентироваться на результат своих действий, хороший или плохой. Такой способ называется обучением с подкреплением. Вот его свойства:

- Есть целевой признак — например прибыль от покупки и продажи акций. В этом смысле всё как в обучении с учителем.

- Нет заранее подготовленного датасета. Учебные данные поступают из среды, в которой учится модель. Пример такой среды — биржа.
- Нет самого учителя — то есть примеров правильных решений. Компьютер вырабатывает их самостоятельно. <font color='green'>Если модель МО — это робот для трейдинга, то она будет запоминать, какие её действия принесли прибыль, а какие — убыток</font>. Поэтому обучение — с подкреплением.

почему бы не обучить робота для трейдинга с помощью регрессии? Ведь он не первый на свете покупает и продаёт акции. Дело в том, что некоторые бизнес-цели требуют анализировать информацию сразу же после того, как она поступила. В торговле на бирже большую роль играют новости. Одни мировые события цикличны, а другие непредсказуемы и происходят один раз, но сильно влияют на рынок. Чтобы прогнозировать котировки акций как можно эффективнее, нужно постоянно обновлять учебные данные и выводы о верных действиях. Обучение с подкреплением не панацея, но регрессия и вовсе здесь бесполезна. Она учит по старым данным.

В обучении с подкреплением пока не выделили классы задач.

##### Обучение без учителя

В обучении без учителя машина ищет не целевой признак одного объекта, а любые закономерности на уровне объектов. Это полезно в задачах кластеризации и снижения размерности.

<font color='red'>Кластеризация</font>. Группировка объектов по их признакам. Модель сама решает по каким. 
С помощью кластеризации можно, например, распределить клиентов по предпочтениям и делать им подходящие предложения.

<font color='red'>Снижение размерности</font>. Бывает, что в признаковом описании указано слишком много свойств объекта — сотни, тысячи. Моделям машинного обучения тяжело делать выводы по такому количеству возможных переменных. Значит, лучше его уменьшить. Специалист по Data Science может сгруппировать признаки вручную, но задача снижения размерности помогает не тратить на это время.

Вот основные черты обучения без учителя:
- Целевого признака нет, потому что задача — выявить закономерности между объектами, а не признаками.

- Примеров правильных решений тоже нет. По ходу обучения компьютер делает выводы сам — без учителя.
- В отличие от обучения с подкреплением, все учебные данные есть изначально. Так же, как в обучении с учителем.

Обощенная схема моделей с учителем, подкреплением, без учителя

##### Тренировочные, валидационные и тестовые данные

После обучения проводят дополнительную настройку модели — первичную валидацию. 

Новые данные нужны для трёх процедур: обучения модели, первичной валидации и тестов. Чтобы обеспечить весь процесс наблюдениями, датасет от заказчика разбивают на три части — тренировочную, валидационную и тестовую выборки.

<font color='red'>Тренировочная выборка.</font> Её также называют обучающей. Это набор объектов, на котором учится модель МО.

<font color='red'>Валидационная выборка</font>. После обучения специалисты часто корректируют модель дополнительными параметрами. Как понять, что добавить в формулу? Нужно дать модели поработать на данных, которые отличаются от тренировочных, и заметить ошибки. Тогда станет ясно, как их исключить. Для этих целей нужна отдельная выборка данных — валидационная. Не все базовые модели требуют донастройки, поэтому валидационную выборку выделяют не всегда.

<font color='red'>Тестовая выборка</font>. Настройка модели похожа на её тест, но всё же нужен отдельный этап тестирования. Дело в том, что параметры модели подбирают под данные валидационной выборки, но как она отработает на других? Чтобы окончательно убедиться в качестве модели машинного обучения, нужна третья порция свежих данных — тестовая выборка.

Деление на тренировочную, валидационную и тестовую выборки в соотношении 3:1:1, как на схеме ниже:



Разбивка данных на выборки

- Первым аргументом передать в функцию train_test_split() входные признаки.

- Вторым аргументом передать целевой признак.
- Третьим аргументом — сообщить размер тестовой выборки, если он отличается от дефолтного в 0.25. Можно указать либо дробное число от 0 до 1, либо целое. В первом случае датасет разделится на доли, а во втором тестовая выборка будет состоять из того количества объектов, которое вы указали целым числом.
- Добавить в функцию именованный аргумент random_state, чтобы при каждом перезапуске блокнота датасет делился на выборки одинаково. Поясним. Функция train_test_split() разбивает наблюдения по частям рандомно — в тестовую выборку всегда будет попадать, к примеру, четверть строк, но из разных мест в таблице. Если перезапустить блокнот, система повторит это действие. Она поделит данные в прежнем соотношении, но в выборки случайным образом попадут другие строки, а значит, модель отработает с новым результатом. В итоге это приведёт к тому, что данные для обучения и тестирования модели пересекутся, и сделать выводы об этих процедурах будет невозможно. Чтобы такого не случилось, нужно зафиксировать деление датасета на выборки по конкретным строкам. Это и делает аргумент random_state.

Теперь посмотрим как справляется модель, которая использует почищенные данные для предсказания

```python
# необходимые импорты
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


# создание константы RANDOM_STATE
RANDOM_STATE = 42


# считывание CSV-файла в датафрейм pandas, сохранение выборок
df = df_estate_clean.dropna()
X = df.drop('last_price', axis=1)
y = df['last_price']
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    random_state=RANDOM_STATE
)

# инициализация и обучение модели
model_lr  = LinearRegression()
model_lr.fit(X_train, y_train)

# предсказанные значения для X_test
predictions = model_lr.predict(X_test)

# вывод первых предсказанного и настоящего значений
print(predictions[0], y_test.head(1))
```

значения предсказанной цены и фактической очень похожи

##### Качество модели МО

<font color='red'>Среднеквадратичная ошибка</font>

Помните, как работает метод наименьших квадратов ошибок? Он считает сумму квадратов всех ошибок среди предсказанных целевых признаков. Сама по себе эта сумма ничего не говорит о качестве модели, ведь одно и то же отклонение на тысячу и миллион строк — совершенно разный результат. 

Но если разделить сумму квадратов ошибок на число объектов прогноза, то получится средний квадрат ошибки. 

Это и есть MSE:

```python
from sklearn.metrics import mean_squared_error 
```

Считаем значение сердеквадратичной ошибки

```python
y_pred, y_true  =  predictions, y_test
mse = mean_squared_error(y_true, y_pred)
mse
```

<font color='red'>Корень среднеквадратичной ошибки</font>

Хорошо, что можно посчитать MSE и не пролистывать мышкой сотни тысяч строк. Но будем честными: значения этой метрики выглядят громоздко. В предсказании цены квартир модель в среднем ошибается на

##### Схема работы с моделью машинного обучения

1) Анализ бизнес-задачи. Сперва всегда нужно обдумать запрос заказчика. Это поможет подобрать вид и задачу машинного обучения так, чтобы достичь цели бизнеса.

2) Сбор данных. Для машинного обучения необходимы данные. Бывает, что датасет предоставляет заказчик, а иногда его приходится собирать самостоятельно. Как вариант, запросить у дата-инженеров.
3) Предобработка данных. Дубликаты, аномалии и пропуски искажают результаты машинного обучения. Обязательно нужно очистить датасет.
4) Исследовательский анализ данных. Хорошее исследование позволит окончательно определиться с базовой моделью МО.
5) Подготовка данных к машинному обучению. Данные готовят по-разному в зависимости от задачи, но в этот этап всегда входят разметка и деление на выборки. Если задача позволяет обучить только одну модель, которую не нужно настраивать, то данные достаточно разделить на тренировочную и тестовую выборки.
6) Обучение модели. Здесь всё ясно: это этап обучения модели на тренировочной выборке данных.
7) Тестирование модели. Когда модель обучена, нужно получить предсказания модели на тестовой выборке и проверить результат с помощью метрик качества.
8) Интерпретация результатов. Специалисты по Data Science выделяют трактование метрик качества и презентацию модели заказчику в отдельный этап.
