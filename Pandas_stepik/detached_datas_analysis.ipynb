Найти разницу между максимальным и минимальным значением цены в каждой категории

```python
import numpy as np 
def diffusion(lst):
    return abs(np.diff([max(lst), min(lst)])[0])

products_df.groupby('CategoryName').agg({'UnitPrice': diffusion }).\
    rename(columns={'UnitPrice': 'delta_price'})
```

Еще проще через lambda

```python
products_df.groupby('CategoryName').agg({'UnitPrice': lambda x: max(x)-min(x) }).\
    rename(columns={'UnitPrice': 'delta_price'})
```

### Присоедние таблиц merge()

Часто бывает, что ключи в таблицах называется по-разному. Например в датафрейме df_c ID клиента - это CustID, а в датафрейме df_o ID клиента - это CustomerID. В таком случае в методе .merge() нужно прописать дополнительные параметры left_on и right_on, где:

left_on - ключ левой таблицы (к которой присоединяем)

right_on - ключ правой таблицы (которую присоединяем)

df_c.merge(df_o, left_on ='CustID', right_on = 'CustomerID')

        Все виды джойнов для merge

how = 'inner' - этот джойн работает по умолчанию. Его можно не прописывать df_с.merge(df_o);

how = 'left'; обозначает, что мы хотим вывести все строки из левой таблицы, независимо от того есть ли соотношение по общему ключу в правой таблице. 

how = 'right'; обозначает, что мы хотим вывести все строки из правой таблицы, независимо от того есть ли соотношение по общему ключу в левой таблице. 

how = 'outer' - часто этот джойн называют full;

how = 'cross'; 




Отдельно по джойну cross - это декартово произведение.

каждая строка из левого датасэта смерджатся с каждой строкой из правого датасэта



Как зовут сотрудника, который оформил больше всего заказов за все время?

По шагам:
1) возьмем две таблицы:
- список сотрудников (личное дело) 
- список заказов за все время (где указано, какой заказ кто из сотридуников обработал)

2) создадим временный датасэт из таблицы заказов - посчитаем количество заказов, которое провел каждй сотрудник

3) смерджим временный датасэт с таблицей личных дел по уникальному номеру ID для каждого сотрудника
4) отфильтруем полученную таблицу - найдем строку с максимальным числом заказов

5) после фильтрации возьмем только столбцы имени и фамилии, возьмем значения этих столбцов

6) вытянем через индексацию строки - и соединим их через джоин

### Замена пустых значений для некоторых строк

достаточно вызвать метод where (условие/фильтр, на что заменить)
- при этом заменятся все строки, которые НЕ попадают под условие/фильтр
- поэтому для того, чтобы заменить только отфильтрованные строки, нужно поставить отрицание фильтра: " ~ "

```python
#заменяем значения столбца ShippedDate для заказов 11076,11077 на Notshipped

orders_df['ShippedDate'] = orders_df['ShippedDate'].\
    where(~(orders_df['OrderID'].isin([11076,11077])), 'Notshipped')        # здесь прописываем на что заменить

orders_df.tail()
```

при этом! 

В столбцах с датами должны храниться только даты. В случае отсутствия дат - должны быть пропущенные значения (NaT, NaN). 

Еще способ замены - через библиотеку numpy и метод replace

```python
import numpy as np
result = orders_df.replace( np.nan,'Not Provided')  

result
```
