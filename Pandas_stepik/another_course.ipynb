Вопрос 10: Сколько заработала фирма на СМС сообщениях? Тарифная сетка:

ВИП - 1 рубль за СМС

Все за 300 - 2 рубля за СМС

Всё за 500 - 3 рубля за СМС

Всё за 800 - 4 рубля за СМС

Корпоративный - 0 рублей за СМС

Всё за 250 (архив) - 5 рублей за СМС

```python
true_sms = df.groupby('tarif', as_index=False).agg({'sms': 'sum'})
sms_dic = {'ВИП': 1, 'Все за 300': 2, 'Всё за 500':3, 'Всё за 800': 4, 'Промо': 7, 'Корпоративный':0, 'Всё за 250 (архив)':5}
earned_on_sms = true_sms.merge(pd.Series(sms_dic).reset_index().rename(columns = {'index':'tarif', 0:'price'}), on='tarif')

earned_on_sms['earned'] = earned_on_sms.sms * earned_on_sms.price
earned_on_sms

```

```python
"""Поступила задача разбить клиентов на группы:

до 25 лет (включительно) - A (англ. буква)
с 26 до 40 лет (включительно) - B (англ. буква)
старше 40 лет - C (англ. буква)"""

def group_users(user):
    if user<=25: return 'A'
    elif 26<=user<=40: return 'B'
    elif user>40: return 'C'

df['group'] = df.age.map(group_users)
df
```

```python
"""Пришла информация, что на сервере в Москве возник сбой и баланс за отправленные смс не списался 
(у пользователей из Москвы). 
Спишите с каждого пользователя из Москвы деньги за смс. 
Количество отправленных смс за которые не списали деньги в столбце sms. Стоимость 1 смс - 3 рубля. 
(обновите колонку balance):"""
temp_list = []
for ind, row in df.iterrows():
    if row.city == 'Moscow':res = row.balance - row.sms *3
    else: res = row.balance
    temp_list.append(res)

deb_funds.balance = pd.Series(temp_list)

deb_funds
```

    ВИП - 1 рубль за СМС
    Все за 300 - 2 рубля за СМС

```python
sms_dic = {'ВИП': 1, 'Все за 300': 2, 
           'Всё за 500':3, 
           'Всё за 800': 4, 
           'Промо': 7, 
           'Корпоративный':0, 
           'Всё за 250 (архив)':5}
```

```python
true_deb_funds = df.copy()
true_deb_funds['price'] = true_deb_funds.tarif.map(lambda x: sms_dic[x])
true_deb_funds['balance'] = np.where(true_deb_funds['city'] == 'Moscow', 
                                true_deb_funds['balance'] - true_deb_funds['sms']* true_deb_funds.price, 
                                true_deb_funds['balance'])

true_deb_funds
```
