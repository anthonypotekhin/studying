```python
# 3D Ð²Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ Ð·Ð°Ð²Ð¸ÑÐ¸Ð¼Ð¾ÑÑ‚Ð¸ Ð±ÐµÐ´Ð½Ð¾ÑÑ‚Ð¸ Ð¾Ñ‚ hs_grad Ð¸ metro_res Ñ Ð´Ð¾Ð±Ð°Ð²Ð»ÐµÐ½Ð¸ÐµÐ¼ Ð¿Ð»Ð¾ÑÐºÐ¾ÑÑ‚Ð¸ Ð½Ð°ÑˆÐµÐ¹ Ñ€ÐµÐ³Ñ€ÐµÑÑÐ¸Ð¾Ð½Ð½Ð¾Ð¹ Ð¼Ð¾Ð´ÐµÐ»Ð¸ (Python):

import statsmodels.formula.api as smf
import pandas as pd
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook

data = pd.read_csv('http://d396qusza40orc.cloudfront.net/statistics/lec_resources/states.csv')

data.head()

lm = smf.ols(formula='poverty ~ metro_res + hs_grad', data=data).fit()

lm.params

def f(x, y):
    return lm.params.Intercept + lm.params.hs_grad * x  + lm.params.metro_res * y

x = data.hs_grad.sort_values()
y = data.metro_res.sort_values()

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.3)
data_below_serf = data[data['poverty'] < f(data['hs_grad'], data['metro_res'])]
data_above_serf = data[data['poverty'] >= f(data['hs_grad'], data['metro_res'])]
ax.scatter(data_below_serf.hs_grad, data_below_serf.metro_res, data_below_serf.poverty, c='r', marker='o')
ax.scatter(data_above_serf.hs_grad, data_above_serf.metro_res, data_above_serf.poverty, c='g', marker='o')
ax.set_xlabel('hs_grad')
ax.set_ylabel('metro_res')
ax.set_zlabel('poverty')
```

```python
# ÐŸÐ°Ñ€Ð°Ð»Ð»ÐµÐ»ÑŒÐ½Ð°Ñ Ð¿Ñ€Ð¾Ð²ÐµÑ€ÐºÐ° Ð½ÐµÐ·Ð°Ð²Ð¸ÑÐ¸Ð¼Ñ‹Ñ… Ð¿ÐµÑ€ÐµÐ¼ÐµÐ½Ð½Ñ‹Ñ… Ð½Ð° Ð¼ÑƒÐ»ÑŒÐºÐ¾Ð»Ð»Ð¸Ð½ÐµÐ°Ñ€Ð½Ð¾ÑÑ‚ÑŒ - Ð¿Ð¾Ð¸ÑÐº Ð¼Ð°ÐºÑÐ¸Ð¼Ð°Ð»ÑŒÐ½Ð¾Ð³Ð¾ Adj R-squared

df = pd.read_csv('http://d396qusza40orc.cloudfront.net/statistics/lec_resources/states.csv')

# ÐÐ°Ð¹Ð´Ñ‘Ð¼ Ñ‚Ð°ÐºÐ¾Ð¹ Ð½Ð°Ð±Ð¾Ñ€ Ñ„Ð¸Ñ‡ÐµÐ¹, Ñ‡Ñ‚Ð¾ R-squared Ð¼Ð°ÐºÑÐ¸Ð¼Ð°Ð»ÑŒÐ½Ð°
feats = {'metro_res', 'white', 'hs_grad', 'female_house'}

# Ð ÐµÐºÑƒÑ€ÑÐ¸Ð²Ð½Ð¾ Ð¿ÐµÑ€ÐµÐ±ÐµÑ€Ñ‘Ð¼ Ð²ÑÐµ ÑÐ¾Ñ‡ÐµÑ‚Ð°Ð½Ð¸Ñ Ð¿ÐµÑ€ÐµÐ¼ÐµÐ½Ð½Ñ‹Ñ…
def brute_force(params: set, formula: str, result: set) -> str:
    if not params:
        result.add(formula.lstrip('+ '))
        return 
    else:
        new_params = params.copy()
        for el in params:
            new_params.remove(el)

            brute_force(new_params, formula + ' + ' + el, result)
            brute_force(new_params, formula, result)

res = set()          
brute_force(feats, '', res)
res.remove('')

# Ð¢ÐµÐ¿ÐµÑ€ÑŒ Ð¿Ð¾ÑÑ‡Ð¸Ñ‚Ð°ÐµÐ¼ Adjusted R-Square Ð´Ð»Ñ ÐºÐ°Ð¶Ð´Ð¾Ð³Ð¾ ÑÐ¾Ñ‡ÐµÑ‚Ð°Ð½Ð¸Ñ Ð¸ Ð²Ñ‹Ð²ÐµÐ´ÐµÐ¼ Ð½Ð° ÑÐºÑ€Ð°Ð½
for formula in res:
    lm = smf.ols(formula='poverty ~ ' + formula, data=df).fit()
    print(f'{lm.rsquared_adj:.2}', formula)
```

```python
# Ð²Ñ‹Ð²ÐµÐ´ÐµÐ¼ Ð¿Ð¾ÑÑ‚Ñ€Ð¾ÐµÐ½Ð¸Ðµ Ð³Ñ€Ð°Ñ„Ð¸ÐºÐ¾Ð² Ð´Ð»Ñ Ð¾Ñ†ÐµÐ½ÐºÐ¸ Ð³ÐµÑ‚ÐµÑ€Ð¾ÑÐºÐµÐ´Ð°ÑÑ‚Ð¸Ñ‡Ð½Ð¾ÑÑ‚Ð¸ Ð¾ÑÑ‚Ð°Ñ‚ÐºÐ¾Ð² Ð¸ Ð¸Ñ… Ð½Ð¾Ñ€Ð¼Ð°Ð»ÑŒÐ½Ð¾ÑÑ‚Ð¸ Ñ€Ð°ÑÐ¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ñ

import pandas as pd
import statsmodels.formula.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats


data = pd.read_csv('http://d396qusza40orc.cloudfront.net/statistics/lec_resources/states.csv')

# ÐšÐ¾ÑÑ„Ñ„Ð¸Ñ†Ð¸ÐµÐ½Ñ‚Ñ‹ ÑƒÑ€Ð°Ð²Ð½ÐµÐ½Ð¸Ñ Ñ€ÐµÐ³Ñ€ÐµÑÑÐ¸Ð¾Ð½Ð½Ð¾Ð¹ Ð¼Ð¾Ð´ÐµÐ»Ð¸
params = sm.ols(formula="poverty ~ metro_res + white + hs_grad", data=data).fit().params
params


# ÐŸÑ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð½Ñ‹Ðµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ (Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ Ð¸Ð· DataFrame -> ÑƒÑ€Ð°Ð²Ð½ÐµÐ½Ð¸Ðµ Ñ€ÐµÐ³Ñ€ÐµÑÐ¸Ð¾Ð½Ð½Ð¾Ð¹ Ð¼Ð¾Ð´ÐµÐ»Ð¸) 
y = params[0] + params[1] * data['metro_res'] + params[2] * data['white'] + params[3] * data['hs_grad'] 
# Ð’Ñ‹Ñ‡Ð¸ÑÐ»ÐµÐ½Ð¸Ðµ Ð¾ÑÑ‚Ð°Ñ‚ÐºÐ° Ñ Ð¿Ð¾Ð¼Ð¾Ñ‰ÑŒÑŽ Ð±Ð¸Ð±Ð»Ð¸Ð¾Ñ‚ÐµÐºÐ¸ statsmodels.formula.api - sm
# ÐžÑÑ‚Ð°Ñ‚ÐºÐ¸ - Ð½Ð° ÑÐºÐ¾Ð»ÑŒÐºÐ¾ y Ð´Ð°Ð»ÐµÐºÐ° Ð¾Ñ‚ Ñ‚ÐµÐ¾Ñ€ÐµÑ‚Ð¸Ñ‡ÐµÑÐºÐ¾Ð¹ Ð¼Ð¾Ð´ÐµÐ»Ð¸
residuals = sm.ols(formula="poverty ~ metro_res + white + hs_grad", data=data).fit().resid
# residuals = data['poverty'] - y - Ñ‚Ð¾ Ð¶Ðµ ÑÐ°Ð¼Ð¾Ðµ
# Ð¡Ð¾Ð·Ð´Ð°ÐµÐ¼ Ð³Ñ€Ð°Ñ„Ð¸ÐºÐ¸
figure, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 6))
# Scatter plot
plt.sca(ax1)
sns.scatterplot(x=y, y=residuals)
sns.lineplot(x=y, y=0, linestyle='--', color='r')
plt.title('ÐÐ½Ð°Ð»Ð¸Ð· Ð¾ÑÑ‚Ð°Ñ‚ÐºÐ¾Ð²')
plt.xlabel('ÐŸÑ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð½Ñ‹Ðµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ')
plt.ylabel('ÐžÑÑ‚Ð°Ñ‚ÐºÐ¸')

#QQ-plot
plt.sca(ax2)
stats.probplot(y, plot=sns.mpl.pyplot)
plt.title('QQ-plot')
```
